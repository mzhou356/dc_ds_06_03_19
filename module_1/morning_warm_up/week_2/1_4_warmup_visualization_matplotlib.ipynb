```python
import matplotlib.pyplot as plt
from IPython.display import Image
import numpy as np
```

```python
from natee_mindy_cliff_CodeWarmup_06_07_19 import df
```

```python
df.head()
```

```python
df = df.loc[:,['Year','Month','Real Average','Interpolated Average']]  # chose the correct cols
```

```python
df.head()
```

```python
# Whenever there is a missing value (-99.99) replace it with the corresponding value at column 4.
df['Real Average'] = np.where(np.isnan(df['Real Average']), df['Interpolated Average'], df['Real Average'])
```

```python
df.head()
```

```python
grouped = df.groupby('Year').mean().reset_index()
grouped.head()
```

```python
plt.figure(figsize = (8,6), facecolor = 'lightgrey')
plt.plot(grouped.Year, grouped['Real Average'], color = 'red')
plt.scatter(df.Year, df['Real Average'], c=df['Month'])
plt.xlim([1975,2020])
plt.ylim([300,420])
plt.yscale("linear")
plt.title('CO2 index by month(dots) & yearly avg (redline)')
plt.savefig('co2_graph.png')
plt.show()
```
