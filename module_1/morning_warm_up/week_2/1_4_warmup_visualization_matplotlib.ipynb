### 3. Using matplotlib we want to create the following figure : 
![Alt Image Text](./plots/CO2_warmup.png)

- Each month is signaled by different colour 
- Add an yearly average (red) (use groupby with mean)
- Notice the starting year on the x-axis
- Notice the range on the Y-axis


```python
grouped = df.groupby('Year').mean().reset_index()
grouped = grouped[grouped.Year >=1980]
df = df[df.Year >=1980]  # chop off at 1980
```

```python
fig, ax = plt.subplots(figsize=(8,6), facecolor = 'lightgrey')
ax.plot(grouped.Year, grouped['Real Average'], color = 'red')
ax.scatter(df.Year, df['Real Average'], c=df['Month'])
ax.set_xlim([1975,2020])
ax.set_ylim([300,420])
ax.set_yscale("linear")
ax.set_title('CO2 index by month(dots) & yearly avg (redline)')
plt.show()
```
